```python
index = 5

velo = np.fromfile("/hdd/kitti/" + gts[index]["point_cloud"]["velodyne_path"], dtype=np.float32).reshape(-1, 4)

plt.figure(figsize=(8,8))
plt.xlim(0, 69.12)
plt.ylim(-39.68, 39.68)
plt.scatter(velo[:, 0], velo[:, 1], s=0.01)

loc = cam2velo(data[index]["location"])
plt.scatter(loc[:, 0], loc[:, 1], label="pred", c="y")
loc = cam2velo(gts[index]["annos"]["location"][gts[index]["annos"]["index"] >= 0])
plt.scatter(loc[:, 0], loc[:, 1], label="gt", c="r", marker="+")
plt.legend()

plt.show()

print(gts[index]["annos"]["name"])
print(gts[index]["annos"]["difficulty"])
print(data[index]["official/3d_0.70/0/bin"])
print(data[index]["official/3d_0.70/1/bin"])
print(data[index]["official/3d_0.70/2/bin"])

plt.title("overlap")
sns.heatmap(data[index]["official/overlaps"][:, gts[index]["annos"]["difficulty"] >= 0])
plt.xlabel("gt")
plt.ylabel("det")
plt.show()

plt.title("overlap > 0.70")
sns.heatmap(data[index]["official/overlaps"][:, gts[index]["annos"]["difficulty"] >= 0] > 0.70)
plt.xlabel("gt")
plt.ylabel("det")
plt.show()

tp = data[index]["official/3d_0.70/2/bin"] == 1
ign = data[index]["official/3d_0.70/2/bin"] == 0
fp = data[index]["official/3d_0.70/2/bin"] == -1


plt.title("confidence scores (3D IoU@0.70)")
plt.bar(np.arange(len(data[index]["score"]))[tp], data[index]["score"][tp], width=1.0, color="orange")
plt.bar(np.arange(len(data[index]["score"]))[ign], data[index]["score"][ign], width=1.0, color="gray")
plt.bar(np.arange(len(data[index]["score"]))[fp], data[index]["score"][fp], width=1.0, color="blue")
plt.legend(["tp", "ignore", "fp"])
plt.show()
```

```python
def show_confidence(index):
    global data
    tp = data[index]["official/3d_0.70/2/bin"] == 1
    ign = data[index]["official/3d_0.70/2/bin"] == 0
    fp = data[index]["official/3d_0.70/2/bin"] == -1
    plt.title("confidence scores (3D IoU@0.70)")
    plt.bar(np.arange(len(data[index]["score"]))[tp], data[index]["score"][tp], width=1.0, color="orange", label="TP")
    plt.bar(np.arange(len(data[index]["score"]))[ign], data[index]["score"][ign], width=1.0, color="gray", label="ignore")
    plt.bar(np.arange(len(data[index]["score"]))[fp], data[index]["score"][fp], width=1.0, color="blue", label="FP")
    plt.legend()
    plt.show()

show_confidence(48)
```
